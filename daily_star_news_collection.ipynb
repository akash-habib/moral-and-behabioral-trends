```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv
from urllib.parse import urljoin

# Function to extract detailed news content from the article page
def get_news_details(article_url, date, title, csv_writer): # Pass csv_writer as an argument
    response = requests.get(article_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract category

    category = ['জাতীয়', 'রাজনীতি', 'অর্থনীতি', 'সারাদেশ', 'প্রথম পাতা', 'শেষের পাতা','দেশ বিদেশ' ,'খবর','বাংলারজমিন','সম্পাদকীয়','অনলাইন']
    category_element = soup.find('h4', class_='sectitle')
    category_element=category_element.get_text(strip=True)
    print(f"Category: {category_element}")

    if category_element in category:
        # Extract the category text
        content_div = soup.find('div', class_="col-sm-10 offset-sm-1 fs-5 lh-base mt-4 mb-5")
        if content_div:
            details=content_div.find_all('p')
            details_paragraphs = " ".join([p.text.strip() for p in details])
            print(details_paragraphs)
            csv_writer.writerow([date, title, article_url, details_paragraphs]) # Now csv_writer is accessible
    return None

# Function to scrape news from a given page URL
def scrape_news_from_page(url, csv_writer, date):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    news_links = soup.find_all('div', class_='row border-bottom py-3 my-3')
    if news_links:
        for cards in news_links:
            link_tag = cards.find('a')
            href = link_tag['href']
            #print(href)
            title= link_tag.text.strip()
            print(title)
            full_link = urljoin("https://mzamin.com/", href)
            details = get_news_details(full_link, date, title, csv_writer) # Pass csv_writer to get_news_details

# Function to scrape news and iterate through pages
def scrape_news_to_csv(base_url, csv_filename, date):
    with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
        csv_writer = csv.writer(file)

        # Write the header only if the file is empty
        if file.tell() == 0:
            csv_writer.writerow(['Date', 'Title', 'Link', 'Details'])
        scrape_news_from_page(base_url, csv_writer, date)

    print("Scraping finished and data saved to CSV.")

def date_to_unix_timestamp(date_str):

    date_obj = datetime.strptime('2024-10-10', '%Y-%m-%d')
    unix_timestamp = int(date_obj.timestamp())-21600
    return unix_timestamp

# Example usage
dates=['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05', '2025-01-06', '2025-01-07', '2025-01-08', '2025-01-09', '2025-01-10', '2025-01-11', '2025-01-12', '2025-01-13', '2025-01-14', '2025-01-15', '2025-01-16', '2025-01-17', '2025-01-18', '2025-01-19', '2025-01-20', '2025-01-21', '2025-01-22', '2025-01-23', '2025-01-24', '2025-01-25', '2025-01-26', '2025-01-27', '2025-01-28', '2025-01-29', '2025-01-30', '2025-01-31']
for date in dates:
    print(date)
    time=date_to_unix_timestamp(date)
    base_url = f'https://mzamin.com/archivedNews.php?data={time}&date={date}'
    csv_filename = 'the_business_standered_march_sept.csv'
    scrape_news_to_csv(base_url, csv_filename,date)
```
