```python
!pip install transformers datasets -q

import pandas as pd
from datasets import Dataset
from transformers import pipeline
import torch

# âœ… Load your data
df = pd.read_csv("/content/all_english_newspaper.csv", on_bad_lines='skip', engine='python')
df = df.head(1000)  # Optional: for testing
df['text'] = df['Title'].fillna('') + '. ' + df['Details'].fillna('')

# âœ… Convert to Hugging Face Dataset
hf_dataset = Dataset.from_pandas(df)

# âœ… Load pipelines on GPU
device = 0 if torch.cuda.is_available() else -1
sentiment_pipe = pipeline("text-classification", model="cardiffnlp/twitter-roberta-base-sentiment", device=device)
emotion_pipe = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", device=device)
zero_shot_pipe = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=device)

# âœ… Labels for Moral Foundations (zero-shot)
moral_labels = [
    "Care/Harm", "Fairness/Cheating", "Loyalty/Betrayal",
    "Authority/Subversion", "Sanctity/Degradation", "Liberty/Oppression"
]

# âœ… Apply Sentiment (fast, batched)
sentiment_results = sentiment_pipe(hf_dataset['text'], batch_size=16, truncation=True, max_length=512)
hf_dataset = hf_dataset.add_column("sentiment", [r['label'] for r in sentiment_results])

# âœ… Apply Emotion
emotion_results = emotion_pipe(hf_dataset['text'], batch_size=16, truncation=True, max_length=512)
hf_dataset = hf_dataset.add_column("emotion", [r['label'] for r in emotion_results])

# âœ… Apply Zero-Shot for Moral Foundations (multi-label)
def get_morals(text):
    out = zero_shot_pipe(text, candidate_labels=moral_labels, multi_label=True)
    top = [label for label, score in zip(out['labels'], out['scores']) if score > 0.5]
    return ', '.join(top) if top else 'None'

# Use .map for efficient processing
hf_dataset = hf_dataset.map(lambda row: {"moral_foundation": get_morals(row["text"])})

# âœ… Convert back to DataFrame and save
final_df = hf_dataset.to_pandas()
final_df.to_csv("labeled_news_all.csv", index=False)
print("âœ… Done! Saved as `labeled_news_all.csv`.")
```
