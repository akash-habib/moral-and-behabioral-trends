```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv
from urllib.parse import urljoin

# Function to extract detailed news content from the article page
def get_news_details(article_url):
    response = requests.get(article_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the content section
    content_div = soup.find('div', class_='post-details-content')
    if content_div:
        details_paragraphs = content_div.find_all('p', class_='rtejustify') or content_div.find_all('p')
        details = " ".join([p.text.strip() for p in details_paragraphs])
        return details if details else "Details not found."
    return "Content not found."

# Function to extract and format the date from the article page
def extract_date(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    date_divs = soup.find_all('div', style="display: block;width:100%;")
    x=""
    for div in date_divs:
        i_tag = div.find('i', class_="fa fa-clock-o")
        if i_tag:
            published_text = div.text.strip().split(":", 1)[1].strip()
            if published_text:
                try:
                    date_format = "%d %b %Y %I:%M %p"
                    published_datetime = datetime.strptime(published_text, date_format)
                    formatted_date = published_datetime.strftime("%Y-%m-%d")
                    x=formatted_date
                    return formatted_date
                except ValueError:
                    return x
    return "Date not found."

# Function to scrape news from a given page URL
def scrape_news_from_page(url, csv_writer):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    #print(soup)

    # Find all news articles on the current page
    news_containers = soup.find_all('div', class_='box-body no-padding')
   # print(news_containers)

    # Loop through each container to extract titles and links
    for container in news_containers:
        # Find all <a> tags inside the container
        links = container.find_all('a')

        # Loop through each link and extract title and href
        for link in links:
            h3 = link.find('h3')
            if h3:
                title =h3.text.strip()  # Get the news title
                href = link['href']  # Get the news link
                full_link = urljoin(url, href)  # Get the full URL
                details = get_news_details(full_link)
                date = extract_date(full_link)

                if details and date:
                    print(f"Collecting news: {title}")
                    print(f"Date: {date}")
                    csv_writer.writerow([date, title, full_link, details])
            else:
                title = link.text.strip()  # Get the news title
                href = link['href']  # Get the news link
                full_link = urljoin(url, href)  # Get the full URL
                details = get_news_details(full_link)
                date = extract_date(full_link)

                if details and date:
                    print(f"Collecting news: {title}")
                    print(f"Date: {date}")
                    csv_writer.writerow([date, title, full_link, details])

# Function to scrape news and iterate through pages
def scrape_news_to_csv(base_url, csv_filename):
    page_number = 2  # Initial page number
    with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
        csv_writer = csv.writer(file)

        # Write the header only if the file is empty
        if file.tell() == 0:
            csv_writer.writerow(['Date', 'Title', 'Link', 'Details'])

        while base_url:
            print(f"Scraping page: {base_url}")
            scrape_news_from_page(base_url, csv_writer)
            page_number += 1  # Increment the page number
            base_url = f'https://bangladeshpost.net/categories/city?page={page_number}'

    print("Scraping finished and data saved to CSV.")

# Example usage
csv_filename = '/content/bangladesh_post_news_data_no_duplicates.csv'
base_url = 'https://bangladeshpost.net/categories/city?page=2'
scrape_news_to_csv(base_url, csv_filename)
```
