```python
!pip install pandas transformers sklearn torch
import pandas as pd
from sklearn.cluster import KMeans
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
from transformers import BertTokenizer, BertModel
import torch
import numpy as np
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()
```

```python
torch.cuda.empty_cache()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
model.to(device)
```

```python
def get_embeddings(texts, batch_size=16):
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i + batch_size]
        # Tokenize and encode the texts
        inputs = tokenizer(batch_texts, padding=True, truncation=True, return_tensors='pt').to(device)  # Move to GPU
        with torch.no_grad():
            outputs = model(**inputs)
        # Use the output of the [CLS] token as the embedding
        embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy()  # Move embeddings back to CPU
        all_embeddings.append(embeddings)
        torch.cuda.empty_cache()  # Clear GPU memory after each batch
    return np.vstack(all_embeddings)  # Stack all embeddings into a single array

# Function to process texts in chunks
def process_in_chunks(texts, chunk_size=1000, batch_size=16):
    all_embeddings = []
    for i in range(0, len(texts), chunk_size):
        chunk = texts[i:i + chunk_size]
        embeddings = get_embeddings(chunk, batch_size=batch_size)
        all_embeddings.append(embeddings)
        torch.cuda.empty_cache()  # Clear GPU memory after each chunk
    return np.vstack(all_embeddings)  # Combine all embeddings into a single array
```

```python
embeddings_before = process_in_chunks(texts_before, chunk_size=1000, batch_size=16)
```

```python
num_topics = 8

kmeans_before = KMeans(n_clusters=num_topics)

kmeans_before.fit(embeddings_before)

labels_before = kmeans_before.labels_
```

```python
before['Topic'] = labels_before

for topic in range(num_topics):
    print(f"Topic {topic} - Before:")
    print(before[before['Topic'] == topic]['Details'].head(10))  # Show top 5 articles
    print("\n")

```

```python
embeddings_after = get_embeddings(texts_after)
```

```python
num_topics = 8
kmeans_after = KMeans(n_clusters=num_topics)
kmeans_after.fit(embeddings_after)
labels_after = kmeans_after.labels_
```

```python
after['Topic'] = labels_after

for topic in range(num_topics):
    print(f"Topic {topic} - After:")
    print(after[after['Topic'] == topic]['Details'].head(10))  # Show top 5 articles
    print("\n")
```
