```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv
from urllib.parse import urljoin

# Function to extract detailed news content from the article page
def get_news_details(article_url):
    response = requests.get(article_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract category
    category = ['জাতীয়', 'রাজনীতি', 'অর্থনীতি', 'সারাদেশ', 'প্রথম পাতা', 'শেষ পাতা', 'খবর','বাংলার মুখ','দ্বিতীয় সংস্করণ','সম্পাদকীয়']
    category_element = soup.find('p', class_='desktopDetailCat marginB30')
    detail = None
    date = None

    if category_element and category_element.a:
        # Extract the category text
        category_text = category_element.a.text.strip()
        print(f"Category: {category_text}")
    else:
        category_text = None


    if category_text and category_element.text.strip() in category:
        content_div = soup.find('div', class_="desktopDetailBody")
        if content_div:
            details_paragraphs = content_div.find_all('p')
            detail = " ".join([p.text.strip() for p in details_paragraphs])
            #return details if details else None

    ptags = soup.find_all('p', class_='desktopDetailPTime color1')

    date_bengali = None
    for p in ptags:
        if 'প্রকাশ' in p.text:
            date_bengali = p.text.replace('প্রকাশ: ', '').split(',')[0].strip()
            print(date_bengali)
            break

    if not date_bengali:
        return None  # If no date found

    # Step 2: Convert Bengali digits to English
    bengali_digits = '০১২৩৪৫৬৭৮৯'
    english_digits = '0123456789'
    digit_map = str.maketrans(bengali_digits, english_digits)

    # Step 3: Convert Bengali month to English
    bengali_months = {
        'জানুয়ারি': 'January',
        'ফেব্রুয়ারি': 'February',
        'মার্চ': 'March',
        'এপ্রিল': 'April',
        'মে': 'May',
        'জুন': 'June',
        'জুলাই': 'July',
        'আগস্ট': 'August',
        'সেপ্টেম্বর': 'September',
        'অক্টোবর': 'October',
        'নভেম্বর': 'November',
        'ডিসেম্বর': 'December'
    }

    for bn_month, en_month in bengali_months.items():
        if bn_month in date_bengali:
            date_bengali = date_bengali.replace(bn_month, en_month)
            break

    # Step 4: Translate digits and format the date
    date_english = date_bengali.translate(digit_map)
    dt = datetime.strptime(date_english.strip(), '%d %B %Y')
    dat = dt.strftime('%Y-%m-%d')
    return detail,dat

# Function to scrape news from a given page URL
def scrape_news_from_page(url, csv_writer):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all news articles on the current page
    news_links = soup.find_all('div', class_='media positionRelative')

    for card in news_links:
        h4 = card.find('h4', class_='title10 marginT2')
        link_tag = card.find('a', class_='linkOverlay')

        if h4 and link_tag:
            title = h4.text.strip()
            href = link_tag['href']
            full_link = urljoin("https://www.jugantor.com", href)

            # Get news details from the article page
            details,date = get_news_details(full_link)
            #if details:
             #   print(f"Collecting news: {title}")
             #   csv_writer.writerow([date, title, full_link, details])
            if details and date:
                print(f"Collecting news: {title}")
                csv_writer.writerow([date, title, full_link, details])

# Function to scrape news and iterate through pages
def scrape_news_to_csv(base_url, csv_filename):
    with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
        csv_writer = csv.writer(file)

        # Write the header only if the file is empty
        if file.tell() == 0:
            csv_writer.writerow(['Date', 'Title', 'Link', 'Details'])

        page_number = 3357
        while base_url:
            print(f"Scraping page: {base_url}")
            scrape_news_from_page(base_url, csv_writer)

            # Check for next page link
            response = requests.get(base_url)
            soup = BeautifulSoup(response.content, 'html.parser')

            next_page = soup.find('a', href=True, rel='next')
            print(next_page)
            if next_page:
                page_number -= 1
                next_href = next_page.get('href')
                base_url = f'https://www.jugantor.com/archive?search=yes&date={date}&page={page_number}'
            else:
                base_url = None  # Stop if no more pages are found

    print("Scraping finished and data saved to CSV.")

# Example usage
dates=['2025-01-01']#, '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05', '2025-01-06', '2025-01-07', '2025-01-08', '2025-01-09', '2025-01-10', '2025-01-11', '2025-01-12', '2025-01-13', '2025-01-14', '2025-01-15', '2025-01-16', '2025-01-17', '2025-01-18', '2025-01-19', '2025-01-20', '2025-01-21', '2025-01-22', '2025-01-23', '2025-01-24', '2025-01-25', '2025-01-26', '2025-01-27', '2025-01-28', '2025-01-29', '2025-01-30', '2025-01-31']
for date in dates:
    base_url = f'https://www.jugantor.com/archive?search=yes&date={date}&page=3357'
    csv_filename = 'news_data_jugantor.csv'
    scrape_news_to_csv(base_url, csv_filename)
```
