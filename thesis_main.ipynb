```python
import pandas as pd
import re
import logging
import uuid
from transformers import pipeline
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import torch
nltk.download('punkt_tab')

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Set up logging
logging.basicConfig(filename='labeling_log.txt', level=logging.INFO, format='%(asctime)s - %(message)s')

# Initialize transformer models
device = 0 if torch.cuda.is_available() else -1  # Use GPU if available
sentiment_classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english', device=device)
emotion_classifier = pipeline('text-classification', model='bhadresh-savani/distilbert-base-uncased-emotion', device=device, return_all_scores=True)

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Expanded keyword dictionaries
moral_foundation_keywords = {
    'Care/Harm': ['suffer', 'harm', 'care', 'protect', 'victim', 'death', 'injury', 'tragedy', 'loss', 'help', 'support', 'pain', 'compassion', 'empathy'],
    'Fairness/Cheating': ['justice', 'fair', 'unfair', 'corruption', 'embezzle', 'fraud', 'equality', 'cheat', 'bribe', 'equity', 'dishonest', 'scandal'],
    'Loyalty/Betrayal': ['loyal', 'betray', 'trust', 'disloyal', 'family', 'group', 'community', 'allegiance', 'treason', 'solidarity'],
    'Authority/Subversion': ['authority', 'leader', 'government', 'law', 'order', 'rebel', 'subvert', 'protest', 'control', 'command', 'defy'],
    'Sanctity/Degradation': ['pure', 'moral', 'immoral', 'degrade', 'violate', 'sacred', 'communal', 'harmony', 'sin', 'defile', 'purity']
}

emotion_keywords = {
    'Anger': ['outrage', 'anger', 'frustrate', 'corrupt', 'injustice', 'betray', 'clash', 'protest', 'rage', 'fury'],
    'Sadness': ['sad', 'tragedy', 'loss', 'death', 'grief', 'suffer', 'suicide', 'mourn', 'sorrow', 'heartbreak'],
    'Fear': ['threat', 'fear', 'danger', 'risk', 'attack', 'panic', 'war', 'conflict', 'anxiety', 'terror'],
    'Joy': ['happy', 'joy', 'success', 'celebrate', 'hope', 'reunite', 'achieve', 'progress', 'delight', 'pride'],
    'Disgust': ['disgust', 'immoral', 'vile', 'repulsive', 'shame', 'degrade', 'abhorrent', 'revolt']
}

# Text preprocessing
def preprocess_text(text):
    if not isinstance(text, str) or not text.strip():
        return ''
    # Lowercase, remove special characters, tokenize
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stop_words]
    return ' '.join(tokens)

def get_sentiment(text):
    """Determine sentiment using transformer model with VADER fallback."""
    if not text:
        logging.info("Empty text for sentiment analysis")
        return 'Neutral', 0.0

    processed_text = preprocess_text(text)
    try:
        result = sentiment_classifier(processed_text[:512])[0]  # Truncate to 512 tokens
        label = 'Positive' if result['label'] == 'POSITIVE' else 'Negative'
        score = result['score']

        # Fallback to VADER if confidence is low
        if score < 0.7:
            vader_scores = analyzer.polarity_scores(text)
            compound = vader_scores['compound']
            if compound > 0.1:
                label = 'Positive'
            elif compound < -0.1:
                label = 'Negative'
            else:
                label = 'Neutral'
            score = compound

        logging.info(f"Sentiment: {label}, Score: {score}")
        return label, score
    except Exception as e:
        logging.error(f"Sentiment error: {e}")
        return 'Neutral', 0.0

def get_emotion(text):
    """Determine emotion using transformer model and keyword augmentation."""
    if not text:
        logging.info("Empty text for emotion analysis")
        return 'Neutral', 0.0

    processed_text = preprocess_text(text)
    try:
        results = emotion_classifier(processed_text[:512])[0]  # Truncate to 512 tokens
        # Map model labels to our emotions
        emotion_map = {
            'joy': 'Joy',
            'sadness': 'Sadness',
            'anger': 'Anger',
            'fear': 'Fear',
            'disgust': 'Disgust',
            'neutral': 'Neutral'
        }

        # Find highest scoring emotion
        max_emotion = max(results, key=lambda x: x['score'])
        emotion = emotion_map.get(max_emotion['label'], 'Neutral')
        score = max_emotion['score']

        # Augment with keyword matching
        text_lower = text.lower()
        keyword_scores = {emotion: 0 for emotion in emotion_keywords}
        for emotion, keywords in emotion_keywords.items():
            for keyword in keywords:
                if keyword in text_lower:
                    keyword_scores[emotion] += 1

        max_keyword_emotion = max(keyword_scores, key=keyword_scores.get)
        if keyword_scores[max_keyword_emotion] > 2 and score < 0.8:
            emotion = max_keyword_emotion
            score = 0.9  # Boost score for strong keyword match

        logging.info(f"Emotion: {emotion}, Score: {score}")
        return emotion, score
    except Exception as e:
        logging.error(f"Emotion error: {e}")
        return 'Neutral', 0.0

def get_moral_foundation(text, tfidf_vectorizer=None, tfidf_matrix=None, index=None):
    """Determine moral foundation using keyword matching and TF-IDF."""
    if not text:
        logging.info("Empty text for moral foundation analysis")
        return 'None', 0.0

    text_lower = text.lower()
    foundation_scores = {foundation: 0 for foundation in moral_foundation_keywords}

    # Keyword matching
    for foundation, keywords in moral_foundation_keywords.items():
        for keyword in keywords:
            if keyword in text_lower:
                foundation_scores[foundation] += 1

    # TF-IDF augmentation
    if tfidf_vectorizer and tfidf_matrix is not None and index is not None:
        tfidf_scores = tfidf_matrix[index].toarray()[0]
        feature_names = tfidf_vectorizer.get_feature_names_out()
        for foundation, keywords in moral_foundation_keywords.items():
            for keyword in keywords:
                if keyword in feature_names:
                    idx = feature_names.tolist().index(keyword)
                    foundation_scores[foundation] += tfidf_scores[idx] * 10  # Scale TF-IDF contribution

    max_foundation = max(foundation_scores, key=foundation_scores.get)
    score = foundation_scores[max_foundation]

    # Rule-based prioritization
    if 'corruption' in text_lower or 'embezzle' in text_lower:
        max_foundation = 'Fairness/Cheating'
        score = max(score, 1.0)
    elif 'death' in text_lower or 'injury' in text_lower:
        max_foundation = 'Care/Harm'
        score = max(score, 1.0)

    result = max_foundation if score > 0 else 'None'
    logging.info(f"Moral Foundation: {result}, Score: {score}")
    return result, score

def label_dataset(input_csv, output_csv):
    """Label the dataset and save to a new CSV."""
    # Read input CSV
    try:
        df = pd.read_csv(input_csv)
    except Exception as e:
        logging.error(f"Error reading CSV: {e}")
        return

    # Handle missing values
    df['Details'] = df['Details'].fillna('')

    # Compute TF-IDF matrix
    processed_texts = [preprocess_text(text) for text in df['Details']]
    tfidf_vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
    tfidf_matrix = tfidf_vectorizer.fit_transform(processed_texts)

    # Apply labeling functions
    sentiments = []
    sentiment_scores = []
    emotions = []
    emotion_scores = []
    moral_foundations = []
    moral_scores = []

    for idx, text in enumerate(df['Details']):
        # Sentiment
        sentiment, s_score = get_sentiment(text)
        sentiments.append(sentiment)
        sentiment_scores.append(s_score)

        # Emotion
        emotion, e_score = get_emotion(text)
        emotions.append(emotion)
        emotion_scores.append(e_score)

        # Moral Foundation
        moral, m_score = get_moral_foundation(text, tfidf_vectorizer, tfidf_matrix, idx)
        moral_foundations.append(moral)
        moral_scores.append(m_score)

    # Add columns to DataFrame
    df['Sentiment'] = sentiments
    df['Sentiment_Score'] = sentiment_scores
    df['Emotion'] = emotions
    df['Emotion_Score'] = emotion_scores
    df['Moral_Foundation'] = moral_foundations
    df['Moral_Score'] = moral_scores

    # Save to output CSV
    try:
        df.to_csv(output_csv, index=False)
        logging.info(f"Labeled dataset saved to {output_csv}")
        print(f"Labeled dataset saved to {output_csv}")
    except Exception as e:
        logging.error(f"Error saving CSV: {e}")

if __name__ == "__main__":
    input_file = 'df_sampled.csv'
    output_file = f'labeled_df_sampled_improved_{uuid.uuid4()}.csv'
    label_dataset(input_file, output_file)
```

```python
import pandas as pd

# Load 1000 labeled data (CSV format)
df = pd.read_csv("/content/combined_df_1_to_1000_with_lebel.csv")

# Encode labels
sent_map = {v: i for i, v in enumerate(df["Sentiment"].unique())}
emo_map = {v: i for i, v in enumerate(df["Emotion"].unique())}
moral_map = {v: i for i, v in enumerate(df["Moral_Foundation"].unique())}

df["Sentiment"] = df["Sentiment"].map(sent_map)
df["Emotion"] = df["Emotion"].map(emo_map)
df["Moral_Foundation"] = df["Moral_Foundation"].map(moral_map)
```

```python
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict
from transformers import XLMRobertaTokenizer

# Split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Convert to Dataset format
dataset = DatasetDict({
    "train": Dataset.from_pandas(train_df),
    "test": Dataset.from_pandas(test_df)
})

# Tokenizer
tokenizer = XLMRobertaTokenizer.from_pretrained("xlm-roberta-base")

def tokenize(batch):
    return tokenizer(batch["Details"], truncation=True, padding="max_length", max_length=256)

dataset = dataset.map(tokenize)
```

```python
import torch.nn as nn
from transformers import XLMRobertaModel

class MultiHeadClassifier(nn.Module):
    def __init__(self, model_name, n_sent, n_emo, n_moral):
        super().__init__()
        self.base = XLMRobertaModel.from_pretrained(model_name)
        self.drop = nn.Dropout(0.1)
        hidden = self.base.config.hidden_size
        self.sent_head = nn.Linear(hidden, n_sent)
        self.emo_head = nn.Linear(hidden, n_emo)
        self.moral_head = nn.Linear(hidden, n_moral)

    def forward(self, input_ids, attention_mask=None):
        out = self.base(input_ids=input_ids, attention_mask=attention_mask)
        cls_out = self.drop(out.last_hidden_state[:, 0, :])
        return {
            "sentiment": self.sent_head(cls_out),
            "emotion": self.emo_head(cls_out),
            "moral": self.moral_head(cls_out)
        }

model = MultiHeadClassifier("xlm-roberta-base", len(sent_map), len(emo_map), len(moral_map))
```

```python
from transformers import Trainer, TrainingArguments
from torch.nn import CrossEntropyLoss

class MultiLabelTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        labels = inputs["labels"]
        outputs = model(input_ids=inputs["input_ids"], attention_mask=inputs["attention_mask"])
        loss_fct = CrossEntropyLoss()
        loss = (
            loss_fct(outputs["sentiment"], labels["Sentiment"]) +
            loss_fct(outputs["emotion"], labels["Emotion"]) +
            loss_fct(outputs["moral"], labels["Moral_Foundation"])
        ) / 3
        return (loss, outputs) if return_outputs else loss

def format_example(example):
    return {
        "input_ids": example["input_ids"],
        "attention_mask": example["attention_mask"],
        "labels": {
            "Sentiment": example["Sentiment"],
            "Emotion": example["Emotion"],
            "Moral_Foundation": example["Moral_Foundation"]
        }
    }

dataset = dataset.map(format_example)

args = TrainingArguments(
    output_dir="trained_model",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=4,
    evaluation_strategy="epoch",
    logging_dir="./logs"
)

trainer = MultiLabelTrainer(
    model=model,
    args=args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    tokenizer=tokenizer
)

trainer.train()
```

```python
# ✅ Step 1: Install libraries
!pip install transformers datasets nltk -q

import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.nn.functional import softmax
from nltk.tokenize import sent_tokenize
from tqdm import tqdm
import nltk
nltk.download("punkt")

# ✅ Step 2: Load and preprocess your data
df = pd.read_csv("/content/drive/MyDrive/all_english_labeled_news.csv")
df=df.loc[25000:]
df = df.dropna(subset=["text"]).reset_index(drop=True)

# ✅ Step 3: Load fast, public DeBERTa NLI model
model_name = "MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name).to(device)

# ✅ Step 4: Truncate each article to 512 tokens
def truncate_to_512_tokens(text):
    encoded = tokenizer.encode(text, max_length=512, truncation=True)
    return tokenizer.decode(encoded, skip_special_tokens=True)

df["text"] = df["text"].apply(lambda x: truncate_to_512_tokens(str(x)))
df = df[df["text"].str.len() > 30].reset_index(drop=True)

# ✅ Step 5: Define moral labels and batch classifier
labels = [
    "Care", "Harm", "Fairness", "Cheating",
    "Loyalty", "Betrayal", "Authority", "Subversion",
    "Sanctity", "Degradation", "Liberty", "Oppression"
]

def make_hypotheses(label_list):
    return [f"This text expresses {label}." for label in label_list]

def classify_batch_manual(batch_texts, batch_size=32):
    all_results = []
    for i in tqdm(range(0, len(batch_texts), batch_size)):
        texts = batch_texts[i:i+batch_size]
        batch_out = []

        for text in texts:
            hypotheses = make_hypotheses(labels)
            premise_list = [text] * len(hypotheses)
            inputs = tokenizer(premise_list, hypotheses, padding=True, truncation=True, return_tensors="pt").to(device)

            with torch.no_grad():
                logits = model(**inputs).logits
            probs = softmax(logits, dim=1)[:, 2]  # entailment class

            label_scores = [(label, float(score)) for label, score in zip(labels, probs) if score > 0.5]
            batch_out.append(label_scores if label_scores else [("None", 1.0)])

        all_results.extend(batch_out)
    return all_results

# ✅ Step 6: Classify and save output
results = classify_batch_manual(df["text"].tolist(), batch_size=32)
df["moral_foundations"] = [[label for label, score in r] for r in results]
df["moral_scores"] = [dict(r) for r in results]

# ✅ Save results
df.to_csv("/content/drive/MyDrive/moral_foundation_labeled_fast.csv", index=False)
print("✅ Done! Results saved to: moral_foundation_labeled_fast.csv")
```
